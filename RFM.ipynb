```
import pandas as pd
import random

# Define lists of possible values
products = ['Manga', 'Books', 'Magazines', 'Stationery', 'Collectibles', 'Light Novels']
book_names = ['Naruto', 'One Piece', 'Attack on Titan', 'Demon Slayer', 'Harry Potter', 'The Hobbit', '1984', 'Tokyo Ghoul']
genres = ['Fantasy', 'Science Fiction', 'Romance', 'Mystery', 'Action', 'Non-fiction', 'Historical Fiction']
places = ['Tokyo, Japan', 'New York, USA', 'London, UK', 'Sydney, Australia', 'Singapore', 'Paris, France', 'Toronto, Canada']
age_groups = ['18-24', '25-34', '35-44', '45-54', '55+']
email_signups = ['Yes', 'No']

# Generate random data
data = {
    'Product': [random.choice(products) for _ in range(300)],
    'Book Name': [random.choice(book_names) for _ in range(300)],
    'Genre': [random.choice(genres) for _ in range(300)],
    'Customer ID': [f'CUST{1000+i}' for i in range(300)],
    'Sales (USD)': [round(random.uniform(5, 200), 2) for _ in range(300)],
    'Demographics (Age Group)': [random.choice(age_groups) for _ in range(300)],
    'Online Sales (USD)': [round(random.uniform(0, 150), 2) for _ in range(300)],
    'Email Sign-up': [random.choice(email_signups) for _ in range(300)],
    'Time Spent in Store (min)': [random.randint(5, 120) for _ in range(300)],
    'Place': [random.choice(places) for _ in range(300)],
    'Price of Product (USD)': [round(random.uniform(5, 100), 2) for _ in range(300)],
    'Number of Items': [random.randint(1, 5) for _ in range(300)]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Calculate Store Sales as the difference between Total Sales and Online Sales
df['Store Sales (USD)'] = df['Sales (USD)'] - df['Online Sales (USD)']

# Show the first 10 rows of the table as an example
print(df.head(10))
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure that the 'Date' column is in datetime format
sales_title['Date'] = pd.to_datetime(sales_title['Date'])

# Extract the month for aggregation purposes
sales_title['Month'] = sales_title['Date'].dt.strftime('%Y-%m')

# Group by Place and calculate total sales for each place
total_sales_by_place = sales_title.groupby('Place')['Sales (USD)'].sum().reset_index()

# Sort places by total sales and select the top 5
top_5_places = total_sales_by_place.sort_values(by='Sales (USD)', ascending=False).head(5)

# Filter the original data to include only the top 5 places
top_5_sales_data = sales_title[sales_title['Place'].isin(top_5_places['Place'])]

# Group by both Place and Month, and sum the sales
top_5_sales_by_month = top_5_sales_data.groupby(['Place', 'Month'])['Sales (USD)'].sum().reset_index()

# Convert 'Month' back to datetime for proper plotting
top_5_sales_by_month['Month'] = pd.to_datetime(top_5_sales_by_month['Month'])

# Set up subplots - one for each place
places = top_5_places['Place'].tolist()  # List of the top 5 places
fig, axes = plt.subplots(5, 1, figsize=(10, 25))  # 5 separate subplots in one column

# Loop over each place and plot its respective data in a separate chart
for i, place in enumerate(places):
    place_data = top_5_sales_by_month[top_5_sales_by_month['Place'] == place]

    sns.lineplot(x='Month', y='Sales (USD)', data=place_data, marker="o", ax=axes[i])

    axes[i].set_title(f'Sales Over Time for {place}')
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel('Sales (USD)')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()



```

```
import numpy as np

# Generate random dates
np.random.seed(0)  # For reproducibility
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2024-01-01')

# Generate a random date for each row in the DataFrame
df['Date'] = pd.to_datetime(np.random.randint(start_date.value, end_date.value, size=len(df)))
print("\nDataFrame with Fake Date Column:")
print(df)
```

```
# Convert PurchaseDate to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Display the DataFrame with the date column
print("DataFrame with Purchase Date:")
print(df)

# Calculate Recency, Frequency, and Monetary Value for RFM analysis
today = df['Date'].max() + pd.Timedelta(days=1)

# Recency (days since last purchase)
recency = (today - df.groupby('Customer ID')['Date'].max()).dt.days

# Frequency (number of purchases)
frequency = df.groupby('Customer ID')['Date'].count()

# Monetary Value (total sales)
monetary = df.groupby('Customer ID')['Sales (USD)'].sum()

# Create RFM DataFrame
rfm = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
}).reset_index()

# Display RFM DataFrame
print("\nRFM DataFrame:")
print(rfm)
```

```
# Histogram of Recency
plt.figure(figsize=(7, 5))
sns.histplot(rfm['Recency'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Recency')
plt.xlabel('Recency (Days Since Last Purchase)')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```

```
# Create Recency segments
bins = [0, 30, 60, 90, 120, 150, np.inf]
labels = ['0-30 days', '31-60 days', '61-90 days', '91-120 days', '121-150 days', '150+ days']
rfm['Recency Segment'] = pd.cut(rfm['Recency'], bins=bins, labels=labels, right=False)

# Bar chart of Recency segments
plt.figure(figsize=(10, 6))
sns.countplot(data=rfm, x='Recency Segment', palette='viridis')
plt.title('Number of Customers by Recency Segment')
plt.xlabel('Recency Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

#NOTE
Segmentation
After analyzing RFM, you can segment customers into different categories, such as:

**Champions**: High Recency, High Frequency, High Monetary – These are your best customers


.**Loyal Customers**: Low Recency, High Frequency, High Monetary – Regular purchasers who need to be kept engaged.


**At-Risk Customers**: High Recency, Low Frequency, Low Monetary – Recent buyers but not frequent or high spenders; may need re-engagement.

**Lost Customers**: High Recency, Low Frequency, Low Monetary – Inactive customers who haven’t purchased recently; often the target for reactivation campaigns.

```
import pandas as pd

# Example DataFrame structure for RFM
data = {
    'Recency': [5, 15, 20, 80, 100],
    'Frequency': [10, 15, 5, 1, 2],
    'Monetary': [200, 1500, 300, 50, 100]
}

rfm = pd.DataFrame(data)

# Define RFM segments
def rfm_segment(row):
    if row['Recency'] <= 30 and row['Frequency'] >= 10 and row['Monetary'] >= 500:
        return 'Champions'
    elif row['Recency'] > 150 and row['Frequency'] >= 10 and row['Monetary'] >= 500:
        return 'Loyal Customers'
    elif row['Recency'] <= 30 and row['Frequency'] < 5 and row['Monetary'] < 500:
        return 'At-Risk Customers'
    elif row['Recency'] > 30 and row['Frequency'] < 5 and row['Monetary'] < 500:
        return 'Lost Customers'
    else:
        return 'Other'

# Apply the segmentation function
rfm['Segment'] = rfm.apply(rfm_segment, axis=1)

# View the segmented DataFrame
print(rfm)
rfm.head(20)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of customers in each segment
segment_counts = rfm['Segment'].value_counts()

# Bar chart of customer segments
plt.figure(figsize=(8, 6))
sns.barplot(x=segment_counts.index, y=segment_counts.values, palette='viridis')
plt.title('Customer Segmentation based on RFM Analysis')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

After segmenting customers based on their RFM (Recency, Frequency, Monetary) scores, you can tailor your marketing and engagement strategies to meet the specific needs of each segment. Here’s a detailed approach for each customer segment:

1. Champions
Characteristics: High Recency, High Frequency, High Monetary

Actions:

Exclusive Offers: Provide them with exclusive discounts, early access to new products, or members-only sales events to maintain their loyalty.
Loyalty Programs: Implement or enhance loyalty programs that reward repeat purchases and encourage referrals.
Personalized Communication: Use personalized emails or messages thanking them for their loyalty and suggesting related products based on their purchase history.
Feedback Solicitation: Ask for feedback on new products or services, giving them a sense of ownership and engagement.
2. Loyal Customers
Characteristics: Low Recency, High Frequency, High Monetary

Actions:

Re-engagement Campaigns: Create targeted campaigns to encourage more frequent purchases, such as reminders about items they might be interested in or discounts on their favorite products.
Content Marketing: Send them valuable content related to their interests, such as how-to guides, tips, or newsletters to keep them engaged.
Special Promotions: Offer limited-time promotions or bundle deals to entice them to make additional purchases.
Personalized Recommendations: Use their purchase history to suggest related products or new arrivals that match their interests.
3. At-Risk Customers
Characteristics: High Recency, Low Frequency, Low Monetary

Actions:

Reactivation Campaigns: Target these customers with special offers or discounts to encourage them to make a purchase.
Surveys and Feedback: Send surveys to understand their experience and why they have not purchased frequently. Use this feedback to make improvements.
Incentives: Consider offering incentives such as free shipping, a discount on their next purchase, or a limited-time offer to motivate them to return.
Education: Provide educational content or tips on how to use the products they previously purchased or related products.
4. Lost Customers
Characteristics: High Recency, Low Frequency, Low Monetary

Actions:

Win-Back Campaigns: Create targeted win-back campaigns with attractive offers or discounts to rekindle their interest.
Personalized Outreach: Reach out personally via email or phone to understand why they stopped shopping and how you can improve their experience.
Feedback Collection: Ask for feedback on their past experiences to identify areas for improvement.
Special Promotions: Consider sending a "We Miss You!" message with a compelling offer to encourage them to return.
Additional Strategies for All Segments
Segmentation-Based Marketing: Use customer segmentation to send tailored marketing messages, making them more relevant and effective.
Cross-Selling and Upselling: Identify opportunities for cross-selling and upselling based on customer behaviors and preferences.
Customer Education: Provide resources, tutorials, and support to help customers make the most of their purchases, enhancing their overall experience.
Social Proof: Use testimonials and reviews to build trust and encourage customers to make purchases, especially for at-risk and lost customers.
